```python
# Weekly Fitness Tracker with Food Recommendation System
```

```python
# ---------------- FUNCTIONS ----------------

# Activity multiplier function

def get_multiplier(activity):
    if activity == "low":
        return 30

    elif activity == "moderate":
        return 33

    elif activity == "high":
        return 36

    else:
        return 33


# Gym split suggestion

def suggest_split(days):
    if days <= 2:
        return "Full Body Split"

    elif days == 3:
        return "Push-Pull-Legs (Modified)"

    elif days == 4:
        return "Upper-Lower Split"

    elif days == 5:
        return "Push-Pull-Legs + Upper"

    else:
        return "Advanced Push-Pull-Legs"

# Muscle frequency recommendation

def muscle_frequency(days):
    if days <= 2:
        return 1

    elif days <= 4:
        return 2

    else:
        return 2.5


# Food recommendation function

def recommend_foods(body_type):

    if body_type == "ectomorph":
        foods = ["Rice", "Chicken", "Peanut Butter", "Milk", "Oats"]

    elif body_type == "mesomorph":
        foods = ["Eggs", "Fish", "Sweet Potato", "Vegetables", "Beef"]

    elif body_type == "endomorph":
        foods = ["Lean Chicken", "Vegetables", "Brown Rice", "Tuna", "Greek Yogurt"]

    else:
        foods = ["Balanced Whole Foods"]

    return foods

# Protein evaluation function

def protein_status(consumed, needed):

    if consumed < needed:
        lacking = needed - consumed
        print("Protein Goal: NOT ACHIEVED")
        print("Protein still needed:", lacking, "g")

    elif consumed == needed:
        print("Protein Goal: EXACTLY ACHIEVED")

    else:
        excess = consumed - needed
        print("Protein Goal: ACHIEVED")
        print("Extra protein consumed:", excess, "g")

# ---------------- MAIN PROGRAM ----------------

print("WEEKLY FITNESS TRACKER")
print("-------------------------")

# User inputs
weight = float(input("Enter your body weight in kg: "))
activity = input("Activity level (low/moderate/high): ").lower()
print("Body Type Guide:")
print("Ectomorph  - Naturally lean, struggles to gain weight")
print("Mesomorph  - Naturally athletic, gains muscle easily")
print("Endomorph  - Gains weight easily, slower metabolism")

body_type = input("Based on the descriptions above, enter your body type: ").lower()
workout_days = int(input("How many days do you train per week?: "))
weekly_budget = float(input("Enter your weekly food budget: "))

calories_per_day = float(input("Average calories consumed per day: "))
protein_per_day = float(input("Average protein consumed per day (g): "))
food_cost_per_day = float(input("Average food spending per day: "))

# Calculations
multiplier = get_multiplier(activity)

maintenance_daily = weight * multiplier
maintenance_weekly = maintenance_daily * 7

weekly_calories = calories_per_day * 7
weekly_protein = protein_per_day * 7
weekly_cost = food_cost_per_day * 7

protein_needed_daily = weight * 1.8
protein_needed_weekly = protein_needed_daily * 7

calorie_difference = weekly_calories - maintenance_weekly

split = suggest_split(workout_days)
frequency = muscle_frequency(workout_days)

# ---------------- OUTPUTS ----------------

print("----- WEEKLY REPORT -----")
print("Estimated weekly maintenance:", maintenance_weekly)
print("Weekly calories consumed:", weekly_calories)
print("Weekly protein consumed:", weekly_protein, "g")
print("Recommended weekly protein:", protein_needed_weekly, "g")

# Calorie analysis
if calorie_difference < -3500:
    print("Status: LARGE CALORIE DEFICIT")

elif calorie_difference < 0:
    print("Status: CALORIE DEFICIT")

elif calorie_difference == 0:
    print("Status: MAINTENANCE")

elif calorie_difference <= 3500:
    print("Status: CALORIE SURPLUS")

else:
    print("Status: LARGE CALORIE SURPLUS")

# Protein analysis
print()
protein_status(weekly_protein, protein_needed_weekly)

# Budget analysis
print("----- BUDGET ANALYSIS -----")

if weekly_cost > weekly_budget:
    excess = weekly_cost - weekly_budget
    print("You exceeded your budget by:", excess)

elif weekly_cost == weekly_budget:
    print("You perfectly matched your budget.")

else:
    savings = weekly_budget - weekly_cost
    print("Remaining budget:", savings)

# ---------------- SAVINGS BANK ----------------

savings_bank = 0

if weekly_cost < weekly_budget:
    savings = weekly_budget - weekly_cost
    savings_bank += savings

    print("Savings this week:", savings)
    print("Total savings in bank:", savings_bank)

elif weekly_cost == weekly_budget:
    print("No savings this week (exact budget used).")

else:
    debt = weekly_cost - weekly_budget
    print("Over budget by:", debt)
    print("No savings added to bank.")


# ---------------- WEIGHT / LOAD TRACKER ----------------

print("----- TRAINING VOLUME TRACKER -----")

# Lists for exercises
exercise_names = []
exercise_weights = []
exercise_sets = []
exercise_reps = []

num_exercises = int(input("How many exercises did you perform?: "))

# Collect exercise information
for i in range(num_exercises):

    name = input("Enter exercise name: ")
    weight_load = float(input("Enter weight/load used: "))
    sets = int(input("Enter number of sets: "))
    reps = int(input("Enter number of reps: "))

    exercise_names.append(name)
    exercise_weights.append(weight_load)
    exercise_sets.append(sets)
    exercise_reps.append(reps)

# Volume calculations
print("----- VOLUME REPORT -----")

for i in range(len(exercise_names)):

    volume = exercise_weights[i] * exercise_sets[i] * exercise_reps[i]

    print("Exercise:", exercise_names[i])
    print("Weight:", exercise_weights[i])
    print("Sets:", exercise_sets[i])
    print("Reps:", exercise_reps[i])
    print("Total Volume:", volume)

# ---------------- SHOPPING RECOMMENDATION ----------------

print("----- POSSIBLE ITEMS TO BUY -----")

items = {
    "Chicken Breast": 250,
    "Eggs": 180,
    "Protein Powder": 1200,
    "Milk": 120,
    "Rice": 300,
    "Oats": 150
}

for item in items:

    if items[item] <= savings_bank:
        print(item, "-", items[item])


# Gym split recommendation
print("----- TRAINING RECOMMENDATION -----")
print("Suggested gym split:", split)
print("Recommended frequency per muscle group:", frequency, "times/week")

# Muscle groups list
muscle_groups = ["Chest", "Back", "Legs", "Shoulders", "Arms"]
workouts = {
    "Chest": ["Push-Ups", "Bench Press", "Chest Fly"],
    "Back": ["Pull-Ups", "Barbell Rows", "Lat Pulldown"],
    "Legs": ["Squats", "Lunges", "Leg Press"],
    "Shoulders": ["Shoulder Press", "Lateral Raise", "Front Raise"],
    "Arms": ["Bicep Curl", "Tricep Pushdown", "Hammer Curl"]
    }

print("Recommended Muscle Group Frequency:")

# Iteration using for loop
for muscle in muscle_groups:
    print(muscle, ":", frequency, "times/week")

# ---------------- TRAINING HISTORY ----------------

print("----- PAST WEEK TRAINING CHECK -----")

# Dictionary for storing frequencies
training_log = {}

# Iteration for user input
for muscle in muscle_groups:
    days = int(input("How many times did you train " + muscle + " last week?: "))
    training_log[muscle] = days

# ---------------- MUSCLE ANALYSIS + WORKOUT SUGGESTION ----------------

print("----- MUSCLE GROUP ANALYSIS -----")

for muscle in muscle_groups:

    if training_log[muscle] < frequency:
        print(muscle, "training frequency is BELOW recommendation.")

        print("Suggested workouts for " + muscle + ":")

        for w in workouts[muscle]:
            print("-", w)

    elif training_log[muscle] == frequency:
        print(muscle, "training frequency is ON TARGET.")

    else:
        print(muscle, "training frequency is ABOVE recommendation.")

# ---------------- SAVINGS BANK ----------------

print("----- SAVINGS BANK -----")

savings_bank = 0

if weekly_cost < weekly_budget:
    savings = weekly_budget - weekly_cost
    savings_bank += savings

    print("Savings added to bank:", savings)
    print("Current savings balance:", savings_bank)

elif weekly_cost == weekly_budget:
    print("No savings added. Exact budget used.")

else:
    print("No savings added due to overspending.")

        
# ---------------- FOOD RECOMMENDATIONS ----------------

print("----- FOOD RECOMMENDATIONS -----")
print("Recommended foods for", body_type)

foods = recommend_foods(body_type)

# Iteration for food list
for food in foods:
    print("-", food)

# Recovery warnings
if workout_days >= 6 and calorie_difference < 0:
    print("Warning: High training frequency with calorie deficit may reduce recovery.")

if protein_per_day < (weight * 1.2):
    print("Warning: Protein intake may be too low for muscle recovery.")

if calories_per_day < (maintenance_daily * 0.7):
    print("Warning: Calorie intake is significantly below maintenance.")
```
